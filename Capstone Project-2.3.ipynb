<H1> Importing libraries </H1>

```python
import requests
import pandas as pd
import numpy as np
```

```python
#Saving the given link in page
page = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
#Taking the request and converting to a usable format
result = requests.get(page)
result_text = result.text
print (result_text)
```

```python
from bs4 import BeautifulSoup
```

```python
#Using BeatifulSoup to understand the tags
soup = BeautifulSoup(result_text,'lxml')
print(soup.prettify())
```

```python
my_table = soup.find("table",{"class":"wikitable sortable"})
```

<H1>Converting the required tags to a list</H1>

```python
Total = my_table.find_all('td')
TotalList=[]
for link in Total:
    TotalList.append(link.get_text())
for i in range(0,len(TotalList),1):
    if (TotalList[i]=='Not assigned' or TotalList[i]=='Not assigned\n'):
        TotalList[i] = ""
TotalList
```

<H2> Making individual lists to add the elements</H2>

```python
PostalCode = []
Borough = []
Neighborhood = []
for i in range(0,len(TotalList),3):
    PostalCode.append(TotalList[i])
    Borough.append(TotalList[i+1])
    Neighborhood.append(TotalList[i+2])
```

```python
PostalCode
```

```python
Borough
```

```python
#Removing the \n from all elements in the Neighborhood
R_Neighborhood = Neighborhood
j=0
for i in Neighborhood:
    R_Neighborhood[j]=i.rstrip("\n")
    j+=1
```

```python
R_Neighborhood
```

```python
#Copying the elements from Borough to Neighborhood in case Neighborhood is empty
for i in range (0,len(Borough),1):
    if (R_Neighborhood[i]==""):
        R_Neighborhood[i]=Borough[i]
R_Neighborhood
```

<H1>Making the Neighborhoods as a single entry if postal codes are same and removing the duplicate Postal Codes</H1>

```python
for i in range(0,len(PostalCode),1):
    for j in range(i+1,len(PostalCode),1):
        if (PostalCode[i] == PostalCode[j]):
            PostalCode[j]=""
            if(R_Neighborhood[i] == R_Neighborhood[j]):
                continue
            else:
                R_Neighborhood[i] = R_Neighborhood[i] + ',' + R_Neighborhood[j]
```

```python
#Adding all the elements to a dataframe
df = pd.DataFrame()
df['PostalCode'] = PostalCode
df['Borough'] = Borough
df['Neighborhood'] = R_Neighborhood
df.head(500)
```

```python
#Finding elements which do not have a Borough
index_no = []
for index, row in df.iterrows():
    if(row["Borough"] == ""):
        index_no.append(index)
index_no
```

```python
#Dropping the rows with no Boroughs
for index, row in df.iterrows():
    for i in range(0,len(index_no),1):
        if(index == index_no[i]):
            df.drop(index = index, inplace = True)
df
```

```python
#Finding elements with no PostalCodes
index_no_new = []
for index, row in df.iterrows():
    if(row["PostalCode"] == ""):
        index_no_new.append(index)
index_no_new
```

```python
#Removing the elements with no postal codes
for index, row in df.iterrows():
    for i in range(0,len(index_no_new),1):
        if(index == index_no_new[i]):
            df.drop(index = index, inplace = True)
df
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            client_id, 
            client_secret, 
            Version, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```
